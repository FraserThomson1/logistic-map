```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
n = 1200
a = np.zeros(shape = (10000,n)) #create 10000 by n matrix
lamb = np.zeros(shape = (n)) #create array of lambda values for different columns
```

```python
for i in range(len(lamb)):
    #increment each lambda value by 0.01 across array
    lamb[i] = 2.81 + i * 0.001
```

```python
#each column in the matrix represents a new series with a different lambda value(120 series)

a[0] = 0.1 #initialise all starting values of a in series to be 0.1
for i in range(1,10000):
    for j in range(n):
        #let (next a) be (lambda * previous a * (1 - previous a))
        a[i][j] = a[i-1][j] * lamb[j] * (1-a[i-1][j])
```

```python
#plot the values of a in different series(different lambda)in the same iteration(in the same row),over a period of iterations.
#x axis of graph ranges from 0 to n representing the n different lambda values
#hence graph shows the value of different series using different lambda at the same iterative step for continuous iterations

n1 = 200 #number of iterations to plot

plt.figure(1)

x = list(range(n))
for i in range(1,n1): #range(1,n1) not range(n1) because a[-0] = a[0] = [-1,-1,-1,...,-1]
    y = a[-i]
    plt.scatter(x,y,s = 0.005)
plt.savefig('logistic_map_2d.pdf')
```

```python
n2 = 150

plt.figure(2)
ax = plt.axes(projection='3d')

# x is the number of the series
# y is the number of the iteration
# z is the values of a in different series

x = list(range(n))

for i in range(1,n2): #range(1,200) not range(200) because a[-0] = a[0] = [-1,-1,-1,...,-1]
    z = a[-i]
    y = i
    ax.scatter3D(x, y, z,s = 1)
plt.savefig('logistic_map_3d.pdf')
```
